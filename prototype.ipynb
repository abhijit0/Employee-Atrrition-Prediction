## Import Necessary Libraries

```python
import numpy as np
import pandas as pd

import sklearn as sk
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split
```

## Import the Training Data and Test Data
- Perform initial preprocessing - Remove Emtpy records 

```python
train_data = pd.read_csv('Train_Dataset.csv')
test_data = pd.read_csv('Test_Dataset.csv')

train_data.replace('nan', np.nan)
train_data.replace('', np.nan)

test_data.replace('nan', np.nan)
test_data.replace('', np.nan)

for column in train_data.columns:
    train_data.dropna(subset=[column], inplace = True)
    
for column in test_data.columns:
    if(str(test_data[column].dtype) == 'object'):
        test_data[column].fillna(test_data[column].mode()[0], inplace=True)
    else:
        test_data[column].fillna(test_data[column].mean(), inplace=True)
    
train_x = train_data.iloc[:, 2:]
train_y = train_data.iloc[:, 1]
test_x = test_data.iloc[:, 1:]

test_x.head()
```

## Seperate Non numeric and numeric attrivutes for further preprocessing

```python
non_num_att = []
num_att = []
for column in train_x.columns:
    if(str(train_data[column].dtype) == 'object'):
        non_num_att.append(column)
    else:
        num_att.append(column)
print(f'Non numeric attributes: {non_num_att}')
print(f'Numeric attributes: {num_att}')
```

```python
for att in non_num_att:
    print(f'{att} {set(train_x[att].values)}')
```

## Data preprocessing - 1
- For Non numerical Features, We replace values with different encoding with same meaning with unique label
- e.g Gneder contains labels as M and Male, we replace every occuring of M with Male. 
- We do the same for Marital Status as well.


```python
train_x['Gender'] = train_x['Gender'].replace('F', 'Female')
#validation_x['Gender'] = validation_x['Gender'].replace('F', 'Female')
train_x['MaritalStatus'] = train_x['MaritalStatus'].replace('M', 'Married')
#validation_x['MaritalStatus'] = validation_x['MaritalStatus'].replace('M', 'Married')

non_att_classes = {}
for att in non_num_att:
    non_att_classes[att] = set(train_x[att].values)
non_att_classes
```

## We Split the training data into training data and validation data to compute accuracy

```python
train_x, validation_x, train_y, validation_y = train_test_split(train_x, train_y, test_size=0.33, random_state=42)
```

## Data preprocessing - 2
- We encode the non numeric features using label encoder
- For numeric features, we convert them to float32 for efficient computation 
- Normalize the numeric features using MinMax Scaler

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
le = preprocessing.LabelEncoder()
scaler = MinMaxScaler()

for att in train_x.columns:
    if(att in non_att_classes.keys()):
        train_x[att] = le.fit_transform(train_x[att].values.reshape(-1,1))
        validation_x[att] = le.fit_transform(validation_x[att].values.reshape(-1,1))
        test_x[att] = le.fit_transform(test_x[att].values.reshape(-1,1))
    else:
        train_x[att] = train_x[att].astype(np.float32)
        validation_x[att] = validation_x[att].astype(np.float32)
        train_x[att] = scaler.fit_transform(train_x[att].values.reshape(-1, 1))
        validation_x[att] = scaler.fit_transform(validation_x[att].values.reshape(-1, 1))
        test_x[att] = scaler.fit_transform(test_x[att].values.reshape(-1, 1))

train_y = train_y.astype(np.float32)
validation_y = validation_y.astype(np.float32)

```

## Model Prediction and evaluation
- We use **Random Forest classifier** for this example
- We evaluate on validation data
- The evaluation metrics is **accuracy**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

clf=RandomForestClassifier(n_estimators=122)
clf.fit(train_x, train_y)

y_pred=clf.predict(validation_x)
print("Accuracy:",metrics.accuracy_score(validation_y, y_pred))
```

## Prediction on test data
- The results are saved into required csv format for each employee id in the test dataset

```python
y_pred_test = clf.predict(test_x)
y_pred_test = y_pred_test.reshape(-1,1)
y_pred_test = y_pred_test.astype(np.int32)
prediction_df = test_data['EmployeeID'].values.reshape(-1,1)
prediction_df = prediction_df.astype(np.int32)
prediction_df = np.concatenate([prediction_df, y_pred_test], axis = 1)
prediction_df = pd.DataFrame(prediction_df, columns=['EmployeeID', 'Attrition'])

print(prediction_df.shape)
```
